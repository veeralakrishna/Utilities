```python
import folium

# create map object centered at New York City
m = folium.Map(location=[40.7128, -74.0060], zoom_start=10)

# add marker for New York City with icon
folium.Marker(
    location=[40.7128, -74.0060],
    popup='New York City',
    icon=folium.Icon(icon='star')
).add_to(m)

# add marker for Boston with icon
folium.Marker(
    location=[42.3601, -71.0589],
    popup='Boston',
    icon=folium.Icon(icon='cloud')
).add_to(m)

# create legend and add to map
legend_html = """
     <div style='position: fixed;
                 bottom: 50px; left: 50px; width: 150px; height: 90px;
                 border:2px solid grey; z-index:9999; font-size:14px;
                 background-color:white; padding: 10px;'>
         <b>Legend</b><br>
         <i class='fa fa-star fa-2x' style='color:blue'></i> New York City<br>
         <i class='fa fa-cloud fa-2x' style='color:green'></i> Boston
     </div>
"""

m.get_root().html.add_child(folium.Element(legend_html))
```

```python
from branca.element import Template, MacroElement

m = folium.Map(location=(30, 20), zoom_start=4)

from branca.element import Template, MacroElement



template = """
{% macro html(this, kwargs) %}

<!doctype html>
<html lang="en">
<head>
  <meta charset="utf-8">
  <meta name="viewport" content="width=device-width, initial-scale=1">
  <title>jQuery UI Draggable - Default functionality</title>
  <link rel="stylesheet" href="//code.jquery.com/ui/1.12.1/themes/base/jquery-ui.css">
  <link rel="stylesheet" href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/5.15.4/css/all.min.css">

  <script src="https://code.jquery.com/jquery-1.12.4.js"></script>
  <script src="https://code.jquery.com/ui/1.12.1/jquery-ui.js"></script>

  <script>
  $( function() {
    var isMinimized = false;

    // Initialize the legend as draggable and set default position to top right
    $( "#maplegend" ).draggable({
        containment: "window",
        start: function (event, ui) {
            $(this).css({
                bottom: "auto",
                right: "auto"
            });
        }
    });

    // Function to minimize or maximize the legend
    $( "#minimize-maximize-button" ).click(function() {
        var legend = $( "#maplegend" );
        var minimizeMaximizeIcon = $(this).children(".fas");

        if (isMinimized) {
            // Maximize the legend
            legend.css({
                bottom: "20px",
                right: "20px",
                left: "auto",
                transform: "none",
                height: "auto"
            });
            minimizeMaximizeIcon.removeClass("fa-plus");
            minimizeMaximizeIcon.addClass("fa-minus");
            isMinimized = false;
        } else {
            // Minimize the legend
            legend.css({
                bottom: "10px",
                left: "50%",
                transform: "translateX(-50%)",
                height: "30px",
                width: "200px"
            });
            minimizeMaximizeIcon.removeClass("fa-minus");
            minimizeMaximizeIcon.addClass("fa-plus");
            isMinimized = true;
        }
    });
});

  </script>
</head>
<body>


<div id='maplegend' class='maplegend'
    style='position: absolute; z-index:9999; border:2px solid grey; background-color:rgba(255, 255, 255, 0.8);
     border-radius:6px; padding: 10px; font-size:14px; right: 20px; top: 20px;'>

<div class='legend-title'>Legend (draggable!)</div>
<div class='legend-scale'>
  <div class='legend-labels' style="display: flex; justify-content: space-between;">
    <div style="display: flex; flex-direction: column; align-items: center;">
        <span style='background:red; opacity:0.7; width: 20px; height: 20px; display: inline-block;'></span>
        <span>Big</span>
    </div>
    <div style="display: flex; flex-direction: column; align-items: center;">
        <span style='background:orange; opacity:0.7; width: 20px; height: 20px; display: inline-block;'></span>
        <span>Medium</span>
    </div>
    <div style="display: flex; flex-direction: column; align-items: center;">
        <span style='background:green; opacity:0.7; width: 20px; height: 20px; display: inline-block;'></span>
        <span>Small</span>
    </div>
  </div>
</div>
<button id="minimize-maximize-button" style="position: absolute; top: 5px; right: 5px; background: none; border: none; cursor: pointer;">
    <i class="fas fa-minus" style="font-size: 16px;"></i>
</button>
</div>

</body>
</html>

<style type='text/css'>
  .maplegend .legend-title {
    text-align: left;
    margin-bottom: 5px;
    font-weight: bold;
    font-size: 90%;
    }
  .maplegend .legend-scale ul {
    margin: 0;
    margin-bottom: 5px;
    padding: 0;
    float: left;
    list-style: none;
    }
  .maplegend .legend-scale ul li {
    font-size: 80%;
    list-style: none;
    margin-left: 0;
    line-height: 18px;
    margin-bottom: 2px;
    }
  .maplegend ul.legend-labels li span {
    display: block;
    float: left;
    height: 16px;
    width: 30px;
    margin-right: 5px;
    margin-left: 0;
    border: 1px solid #999;
    }
  .maplegend .legend-source {
    font-size: 80%;
    color: #777;
    clear: both;
    }
  .maplegend a {
    color: #777;
    }
</style>
{% endmacro %}"""



macro = MacroElement()
macro._template = Template(template)

m.get_root().add_child(macro)

m
```
