```python
def plotly_linechart(df,
                     timestamp,
                     cols,
                     rangeslider_visible=False,
                     showlegend=False,
                     showspikes=False,
                     buttons=False,
                     hovermode="x",
                     title_text="title"):


    fig = go.Figure()

    for i in cols:
        fig.add_trace(go.Scatter(x=df[timestamp],
                                 y=df[i],
                                 mode='lines', # [lines, markers]
                                 name=i,
                                 line=dict(
                                    # color=color_map[i],
                                    width=3),
                                 connectgaps=False,
                                 customdata = [i],
                                 hovertemplate="<br>".join(
                                                            [
                                                                # "Timestamp: %{x}",
                                                                "Value: %{y}"
                                                                ]
                                                           )
                                 )
        )


    if buttons:
        # Add dropdown buttons
        buttons = []
        i = 0

        for col in cols:
            args = [False] * len(cols)
            args[i] = True

            #create a button object for the col we are on
            button = dict(label = col,
                        method = "update",
                        args=[{"visible": args}])

            #add the button to our list of buttons
            buttons.append(button)

            #i is an iterable used to tell our "args" list which value to set to True
            i+=1

        show_all = dict(label="All",
                        method="update",
                        args=[{"visible": [True]*len(cols)}])

        buttons.insert(0, show_all)

        fig.update_layout(updatemenus=[dict(active=0,
                                            type="dropdown",
                                            buttons=buttons,
                                            x = 0,
                                            y = 1.1,
                                            xanchor = 'left',
                                            yanchor = 'bottom'),
                                    ])



    fig.update_layout(
        # Figure title styling
        title= dict(
            text=title_text,
            x=0.6,
            y=0.95,
            font=dict(
                family="Arial",
                size=20,
                color='#000000'
                )
            ),
        xaxis=dict(
            showline=True,
            showgrid=False,
            showticklabels=True,
            linecolor='rgb(158, 152, 152)',
            linewidth=3,
            ticks='outside',
            tickfont=dict(
                family='Arial',
                size=12,
                color='rgb(82, 82, 82)',
            ),
        ),
        yaxis=dict(
            showgrid=False,
            zeroline=False,
            showline=True,
            linecolor='rgb(158, 152, 152)',
            linewidth=3,
            showticklabels=True,
        ),

        hoverlabel=dict(
                bgcolor="rgba(58, 71, 80, 0.1)",
                font_size=16,
                font_family="Rockwell"
                ),

        hovermode=hovermode,

        margin=dict(
            l=0,
            b=0,
            r=0,
            t=0,
        ),
        showlegend=showlegend,
        legend=dict(
                    orientation="h",
                    yanchor="bottom",
                    y=-0.3,
                    xanchor="right",
                    x=1,
                    bgcolor="LightSteelBlue",
                    bordercolor="Black",
                    borderwidth=2,
                ),

        plot_bgcolor='white'
    )

    fig.update_xaxes(
        rangeslider_visible=rangeslider_visible,
        rangeselector=dict(
            buttons=list([
                dict(count=1, label="1d", step="day", stepmode="backward"),
                dict(count=1, label="1m", step="month", stepmode="backward"),
                dict(count=6, label="6m", step="month", stepmode="backward"),
                dict(count=1, label="YTD", step="year", stepmode="todate"),
                dict(count=1, label="1y", step="year", stepmode="backward"),
                dict(step="all")
            ])
        )
    )

    if showspikes:

        fig.update_xaxes(showspikes=True)
        fig.update_yaxes(showspikes=True)

    fig.show()
```

```python
plotly_linechart(df, timestamp, cols, showspikes=True)
```

```python
from plotly.subplots import make_subplots
import pandas as pd


def plotly_line_Charts_facet(df, cols, title_text="Sensor Tag trends for <SensorGroup>"):

    ncols = len(cols)


    # subplot setup
    fig = make_subplots(
        rows=ncols,
        cols=1,
        vertical_spacing=0.025,
        shared_xaxes=True,
        # x_title="Timestamp",
        # subplot_titles=tuple(cols)
        )

    # Subplot title font resize
    # for i in fig['layout']['annotations']:
    #     i['font'] = dict(size=10,color='#ff0000')

    for i, col in enumerate(cols, start=1):
        fig.add_trace(go.Scatter(
            x=df[timestamp],
            y=df[col].values,
            name=col),
            row=i,
            col=1
            )


    fig.update_layout(
        # Figure title styling
        title= dict(
            text=title_text,
            x=0.5,
            y=0.99,
            font=dict(
                family="Arial",
                size=20,
                color='#000000'
                )
            ),
        xaxis=dict(
            rangeselector=dict(
                buttons=[
                dict(count=1,
                    label="1d",
                    step="day",
                    stepmode="backward"),
                dict(count=1,
                    label="1m",
                    step="month",
                    stepmode="backward"),
                dict(count=6,
                    label="6m",
                    step="month",
                    stepmode="backward"),
                dict(count=1,
                    label="YTD",
                    step="year",
                    stepmode="todate"),
                dict(count=1,
                    label="1y",
                    step="year",
                    stepmode="backward"),
                dict(step="all")
            ]),
            type="date"),#end xaxis  definition

        hoverlabel=dict(
            bgcolor="LightSteelBlue", #"rgba(58, 71, 80, 0.1)",
            font_size=16,
            font_family="Rockwell"
            ),

        margin=dict(
            l=0,
            b=0,
            r=0,
            t=0,
        ),
        showlegend=True,
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=-0.2,
            xanchor="right",
            x=1,
            bgcolor="white",
            bordercolor="Black",
            borderwidth=2
            ),

            plot_bgcolor='white',

            height=ncols*0.99*100
        )


    fig.update_xaxes(
        showline=True,
        linewidth=1,
        linecolor='darkgray',
        mirror=True,
        ticks='inside',
        showspikes=True,
        spikemode='across',
        spikesnap='cursor',
        tickcolor= "black"
        )

    fig.update_yaxes(
        showline=True,
        linewidth=1,
        linecolor='darkgray',
        mirror=True,
        ticks='outside',
        side='left',
        showspikes=True,
        spikemode='across',
        spikesnap='cursor'
        )

    fig.update_layout(
        font_size=10,
        hovermode='x unified',
        dragmode='zoom',
        spikedistance = 3
        )

    # Global X-axis Title
    # fig.update_layout(
    #     annotations=[
    #         go.layout.Annotation(
    #             {
    #                 'showarrow': False,
    #                 'text': "Timestamp",
    #                 'x': 0.1,
    #                 'xanchor': 'center',
    #                 'xref': 'paper',
    #                 'y': 0,
    #                 'yanchor': 'top',
    #                 'yref': 'paper',
    #                 'yshift': -30,

    #                 "font": dict(
    #                     # family="Courier New, monospace",
    #                     size=18,
    #                     # color="#ffffff"
    #                     ),

    #             })
    #         ]
    #     )



    fig.show()
```
